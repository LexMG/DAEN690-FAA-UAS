## Latitude/Longitude Conversions

```python
def dms_to_dd (loc):
    '''
    This function takes in a latitude or longitude coordinate in Degrees Minutes Seconds Format and 
    outputs that location in Decimal Degrees Format
    
    @param loc: a str indicating a latitude or longitude (in Degrees, Minutes, Seconds Format)
    
    output: a float describing the lat/long location in Decimal Degrees Format
    '''
    if type(loc) == float or loc == '':
        dd = loc
        
    else:
        loc = loc.strip()
        #loc = loc.replace('', '00-00-00.0000N')
        d, m, s = map(float, loc[:-1].split('-'))
        h = loc[-1]
    
        dd = float(d) + float(m)/60 + float(s)/(60*60);
        if h == 'S' or h == 'W':
            dd *= -1
    
    return dd;
```

```python
# Calls dms_to_dd() function to change all latitudes and longitudes in Airports dataset to Decimal
# Degrees Format

long_list = airports_final['LONGITUDE'].tolist()
lat_list = airports_final['LATITUDE'].tolist()

new_lat = []
new_long = []

for i in range(len(long_list)):
    new_long.append(dms_to_dd(long_list[i]))
    new_lat.append(dms_to_dd(lat_list[i]))

# Replace df columns with new latitude/longitude values (Decimal Degrees Format)
airports_final['LATITUDE'] = new_lat
airports_final['LONGITUDE'] = new_long
```
