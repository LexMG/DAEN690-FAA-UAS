### This notebook links airports dataset with runways dataset 

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Read in airports dataset
airportC = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/airports_cleaned.csv')
airportC = airportC.drop(['ICAO_ID'], axis=1)
airportC
```

```python
# Read in runways dataset
runwayC = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/runways_cleaned.csv')
runwayC
```

```python
# Merge datasets on IDENT
airports_runways = pd.merge(runwayC, airportC, on='IDENT', how='left')
airports_runways
```

```python
# Export to csv file
airports_runways.to_csv('Airports&Runways.csv', index = False)
```
