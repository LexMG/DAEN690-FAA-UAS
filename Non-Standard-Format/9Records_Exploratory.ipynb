```python
# Import libraries
import pandas as pd
import numpy as np
import geopy
from geopy.distance import geodesic
```

```python
# Read in csv file
nonstd = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/Non-std-9Remaing.csv')
nonstd
```

```python
# Extract IDENTs
uas_loc = nonstd['UAS_LOC'].tolist()
uas_airport = []

for i in range(len(uas_loc)):
    airport = uas_loc[i].split(' ')[-1]
    
    if len(airport) <= 4:
        uas_airport.append(airport)
    else:
        trim_air = airport[-7:-2]
        uas_airport.append(trim_air)
        
uas_airport_df = pd.DataFrame()
uas_airport_df['ICAO_ID'] = uas_airport
uas_airport_df
```

```python
# Read in airports dataset
airports = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/airports_cleaned.csv')
```

```python
# Merge dataframe with airports dataset on ICAO_ID
uas_air_loc= pd.merge(uas_airport_df, airports, on='ICAO_ID', how='left')
uas_air_loc
```

```python
# Append neeeded columns
add_col = nonstd['REMARKS'].tolist()
uas_air_loc['REMARKS'] = add_col
add_col1 = nonstd['UAS_LOC'].tolist()
uas_air_loc['UASLOCATION'] = add_col1
uas_air_loc
```

```python
# Find all records that did not match
uas_air_loc1 = uas_air_loc[uas_air_loc['NAME'].isnull()]
uas_air_loc1
```

```python
# Export records to csv file and send to Lex for further investigation
uas_air_loc1 = uas_air_loc1.drop(['NAME', 'LATITUDE', 'LONGITUDE', 'IDENT', 'ICAO_ID', 'UASLOCATION'], axis=1)
uas_air_loc1 = pd.merge(uas_air_loc1, nonstd, on='REMARKS', how='left')
uas_air_loc1.to_csv('Non-std-6Remaining.csv', index = False)
```

```python
# Find matching records
uas_air_loc = uas_air_loc[uas_air_loc['NAME'].notnull()]
uas_air_loc.to_csv('Non-std-3ICAO.csv', index = False)
uas_air_loc = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/Non-std-3ICAO.csv')
uas_air_loc
```

```python
# Create dictionary for bearings and their degrees
bearing_deg = {
    'N' : 0,
    'NNE' : 23,
    'NE' : 45,
    'ENE' : 68,
    'E' : 90,
    'ESE' : 113,
    'SE' : 135,
    'SSE' : 158,
    'S' : 180,
    'SSW' : 203,
    'SW' : 225,
    'WSW' : 248,
    'W' : 270,
    'WNW' : 293,
    'NW' : 315,
    'NNW' : 338
}

# Separate distances and bearings into two columns
uas_loc = uas_air_loc['UASLOCATION'].tolist()
distances = []
bearings = []

for i in range(len(uas_loc)):
    split = uas_loc[i].split(' ')[0:3]

    if 'NM' in split[0]: #if there is no space between the distance and NM (i.e. 3NM)
        distance_nm = split[0][0:len(split[0]) - 2]
        bearing_abrev = split[1].replace('of', '')
        
    else: 
        distance_nm = split[0][2:len(split[0])]
        bearing_abrev = split[2].replace('of', '')
    
    distances.append(distance_nm[0:])
    bearings.append(bearing_abrev)

# Create DataFrame to store UAS Location Information
new_df = pd.DataFrame(uas_air_loc['REMARKS'])
new_df['UAS_LOC'] = uas_air_loc['UASLOCATION']
new_df['ICAO_ID'] = uas_air_loc['ICAO_ID']
new_df['Distance'] = distances
new_df['Bearing'] = bearings

new_df
```

```python
# Convert distances to KM and bearing to degrees
new_df['Distance'] = pd.to_numeric(new_df['Distance'],errors = 'coerce')
dist_kilo = []

for i in range(len(new_df)):
    distanceKilo = new_df['Distance'][i] * 1.852 # converting NM to kilometers
    dist_kilo.append(distanceKilo)

bearDegree = pd.DataFrame(new_df['Bearing'])
bearDegree = bearDegree.replace({'Bearing': bearing_deg})

new_df['Distance_Kilometers'] = dist_kilo
new_df['Bearing_Degrees'] = bearDegree
new_df['Airport_Latitude'] = uas_air_loc['LATITUDE']
new_df['Airport_Longitude'] = uas_air_loc['LONGITUDE']

new_df2 = pd.merge(new_df, airports, on='ICAO_ID', how='left')
new_df2 = new_df2.drop(['Airport_Latitude', 'Airport_Longitude'], 1)
new_df2 = new_df2.rename({'LONGITUDE': 'Airport_Longitude', 'LATITUDE': 'Airport_Latitude'}, axis=1)
new_df2
```

```python
# Merge, drop and rename columns
new_df2 = pd.merge(new_df2, airports, on='ICAO_ID', how='left')
new_df2 = new_df2.drop(['Airport_Latitude', 'Airport_Longitude'], 1)
new_df2 = new_df2.rename({'LONGITUDE': 'Airport_Longitude', 'LATITUDE': 'Airport_Latitude'}, axis=1)
new_df2
```

```python
# Calculate UAS lat/long
uas_lat = []
uas_long = []

for i in range(len(new_df2)):
    lat_airport = pd.to_numeric(new_df2['Airport_Latitude'][i])
    long_airport = pd.to_numeric(new_df2['Airport_Longitude'][i])
    b = pd.to_numeric(new_df2['Bearing_Degrees'][i])
    d = pd.to_numeric(new_df2['Distance_Kilometers'][i])
    
    origin = geopy.Point(lat_airport, long_airport)
    destination = geodesic(kilometers=d).destination(origin,b)
    
    lat2, lon2, = destination.latitude, destination.longitude
    
    uas_lat.append(lat2)
    uas_long.append(lon2)

# Append UAS Lat/Long information to DataFrame
new_df2['UAS_Latitude'] = uas_lat
new_df2['UAS_Longitude'] = uas_long
new_df2
```

```python
# Export records to csv file
new_df2 = new_df2.drop(['Distance','Bearing','Distance_Kilometers','Bearing_Degrees','ICAO_ID', 'NAME_y'], 1)
new_df2 = new_df2.rename({'Airport_Longitude': 'Origin_Longitude', 'Airport_Latitude': 'Origin_Latitude', 'IDENT_x':'IDENT','NAME_x':'NAME'}, axis=1)
new_df2 = pd.merge(new_df2, nonstd, on='REMARKS', how='left')
new_df2 = new_df2.drop(['IDENT_y','IDENT_y','UAS_LOC_y'], 1)
new_df2 = new_df2.rename({'IDENT_x':'IDENT'}, axis=1)
new_df2.to_csv('COMPLETED-Non-std-3ICAO.csv', index = False)
```
