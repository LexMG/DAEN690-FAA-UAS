This Jupyter Notebook links the two supplementary datasets: airports_cleaned.csv (prepared by Grace) and rwy.csv (prepared by Eric) for use in Identifying latitude/longitude coordinates for Runways at specific airports.

*This script contains the following process:*
1. Rename to 'GLOBAL_ID' field from airports_cleaned.csv to 'properties.AIRPORT_ID' for merge purposes
2. Merge airports_cleaned.csv with rwy.csv on 'properties.AIRPORT_ID' field

***NOTE:*** *There are 515 records within the rwy.csv dataset for which there is no associated airport in airports_cleaned.csv*

```python
import pandas as pd
import numpy as np
```

```python
# LINKING airports_cleaned.csv and rwy.csv
airportC = pd.read_csv('C:/Users/grace/OneDrive/Desktop/GMU/DAEN690/airports_cleaned.csv')
airportC.rename(columns={'GLOBAL_ID': 'properties.AIRPORT_ID'}, inplace=True)
airportC
```

```python
runwayC = pd.read_csv('C:/Users/grace/OneDrive/Desktop/GMU/DAEN690/rwy.csv')
runwayC
```

```python
airports_runways = pd.merge(runwayC, airportC, on='properties.AIRPORT_ID', how='left')
```

```python
airports_runways = airports_runways[['properties.GLOBAL_ID', 'properties.AIRPORT_ID', 'properties.DESIGNATOR', 'Runway.Latitude', 'Runway.Longitude', 'IDENT', 'NAME', 'LATITUDE', 'LONGITUDE', 'ICAO_ID']]
```

```python
airports_runways
```

```python
# Export final dataframe (EXCLUDING NULL BEARING INFO) to .csv file
airports_runways.to_csv('airports_runways_linked.csv', index = False)
```
