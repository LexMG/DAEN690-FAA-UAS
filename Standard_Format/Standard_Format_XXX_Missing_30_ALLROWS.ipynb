#### Get complete Runway information from regex and include 'REPORTINGFACILITY' field for link to runways/airports supplemental datasets

```python
# Create List that contains each Standard Remark, and the Heading/Direction
# information contained in each remark

remark_uas_loc = []
remarks = runways14['REMARKS']

# regular expression for any heading/direction
rwy_regex = '\.?[0-9]\.?[0-9]*[0-9]*\s?NM* [N|S|E|W|NW|NE|SW|SE|SSE|SSW|SNE|SNW|NNE|NSE|NNE|NNW|WSW|WNW|WSE|WNE|ENE|ESE|ESW|ENW]*\s?of?\s?[A-Z][A-Z][A-Z][A-Z]?\s?[0-9]+'

# Loop through all remarks and search for the heading/direction regex above
for i in range(len(remarks)):
    rwy = re.findall(rwy_regex, remarks[i])
    remark_uas_loc.append(remarks[i])
    remark_uas_loc.append(rwy)

# Split Remarks and Heading/Directions into two seperate lists and create
# pandas dataframe
remark = []
uas_loc = []

for i in range(0, len(remark_uas_loc), 2):
    remark.append(remark_uas_loc[i])
    uas_loc.append(remark_uas_loc[i+1])

remark_uas_loc_df = pd.DataFrame()
remark_uas_loc_df['REMARKS'] = remark
remark_uas_loc_df['UAS_Location_Runways'] = uas_loc
remark_uas_loc_df['REPORTINGFACILITY'] = runways14['REPORTINGFACILITY']
```

```python
remark_uas_loc_df
```

### Separate UAS_Location_Runways field into Distance (NM) and bearing abbreviation

We will also convert Nautical Miles (NM) to Kilometers and all bearing abbreviations to their respective degrees

```python
# Create Dictionary of Bearings and Respective Degrees
bearing_deg = {
'N' : 0,
'NNE' : 23,
'NE' : 45,
'ENE' : 68,
'E' : 90,
'ESE' : 113,
'SE' : 135,
'SSE' : 158,
'S' : 180,
'SSW' : 203,
'SW' : 225,
'WSW' : 248,
'W' : 270,
'WNW' : 293,
'NW' : 315,
'NNW' : 338
}

uas_loc = remark_uas_loc_df['UAS_Location_Runways']
distances = []
bearings = []

for i in range(len(uas_loc)):
    split = uas_loc[i][0].split(' ')[0:3]

    if 'NM' in split[0]: #if there is no space between the distance and NM (i.e. 3NM)
        distance_nm = split[0][0:len(split[0]) - 2]
        bearing_abrev = split[1].replace('of', '')

    else: 
        distance_nm = split[0]
        bearing_abrev = split[2].replace('of', '')

    distances.append(float(distance_nm))
    bearings.append(bearing_abrev)

# Create DataFrame to store UAS Location Information
new_df = pd.DataFrame(remark_uas_loc_df['REMARKS'])
new_df['UAS_Location_Runways'] = remark_uas_loc_df['UAS_Location_Runways']
new_df['IDENT'] = remark_uas_loc_df['REPORTINGFACILITY']
new_df['Distance_NM'] = distances
new_df['Bearing'] = bearings

# Convert all distance from NM to kilometers and all bearing abbreviations to their associated degrees
dist_kilo = []

for i in range(len(new_df)):
    distanceKilo = new_df['Distance_NM'][i] * 1.852 # converting NM to kilometers
    dist_kilo.append(distanceKilo)

bearDegree = pd.DataFrame(new_df['Bearing'])
bearDegree = bearDegree.replace({'Bearing': bearing_deg})

for i in range(len(bearDegree)):
    if bearDegree['Bearing'][i] == '':
        bearDegree.replace('', np.nan, inplace = True)

new_df['Distance_Kilometers'] = dist_kilo
new_df['Bearing_Degrees'] = bearDegree
```

```python
rwy_linked2.rename(columns = {'Runway.Latitude': 'Runway_Longitude', 'Runway.Longitude' : 'Runway_Latitude'}, inplace = True)
```

### Calculate UAS Latitude/Longitude Information using geopy library

```python
uas_lat = []
uas_long = []


for i in range(len(rwy_linked2)):
    lat_rwy = pd.to_numeric(rwy_linked2['Runway_Latitude'][i])
    long_rwy = pd.to_numeric(rwy_linked2['Runway_Longitude'][i])
    b = pd.to_numeric(rwy_linked2['Bearing_Degrees'][i])
    d = pd.to_numeric(rwy_linked2['Distance_Kilometers'][i])

    origin = geopy.Point(lat_rwy, long_rwy)
    destination = geodesic(kilometers=d).destination(origin,b)

    lat2, lon2, = destination.latitude, destination.longitude

    uas_lat.append(lat2)
    uas_long.append(lon2)

# Append UAS Lat/Long information to DataFrame
rwy_linked2['UAS_Latitude'] = uas_lat
rwy_linked2['UAS_Longitude'] = uas_long
```
