```python
# Search for FAA Airport Identifiers within each of the remaining 1552 OTHER Standard Format Remarks (THAT DO NOT 
                                                                                            # REFERENCE A RUNWAY)
ident_index = []

for i in range(len(unkn_remarks)):
    for ident in faa_ident:
        if ident in unkn_remarks[i]:
            ident_index.append(unkn_remarks[i])
            ident_index.append(ident)
```

```python
ident_index
```

```python
# Split Remarks and Heading/Directions into two seperate lists and create
# pandas dataframe
faa_identifier = []
remark = []

for i in range(0, len(ident_index), 2):
    faa_identifier.append(ident_index[i+1])
    remark.append(ident_index[i])

ident_index_df = pd.DataFrame()
ident_index_df['REMARKS'] = remark
ident_index_df['FAA_IDENT'] = faa_identifier
```

```python
ident_index_df
```

```python
unkn_std_format_FAAIDENT = ident_index_df.groupby(['REMARKS'])['FAA_IDENT'].apply(','.join).reset_index()
unkn_std_format_FAAIDENT
```

```python
# Trying to fix the FAA Identifier hitting on things that are not really referencing an FAA Identifier
# Create List that contains each Standard Remark, and the Heading/Direction
# information contained in each remark

remark_uas_loc = []
remarks = unkn_std_fmt['REMARKS']

# regular expression for any heading/direction
headir_regex = 'o?f?\s?[A-Z][A-Z][A-Z]?[A-Z]\.'

# Loop through all remarks and search for the heading/direction regex above
for i in range(len(remarks)):
    head_dir = re.findall(headir_regex, remarks[i])
    remark_uas_loc.append(remarks[i])
    remark_uas_loc.append(head_dir)

# Split Remarks and Heading/Directions into two seperate lists and create
# pandas dataframe
remark = []
uas_loc = []

for i in range(0, len(remark_uas_loc), 2):
    remark.append(remark_uas_loc[i])
    uas_loc.append(remark_uas_loc[i+1])

remark_uas_loc_df = pd.DataFrame()
remark_uas_loc_df['REMARKS'] = remark
remark_uas_loc_df['UAS Location'] = uas_loc
```

```python
remark_uas_loc_df
```

```python
# Get list of UAS locations from the above dataframe
uas_loc = remark_uas_loc_df['UAS Location'] 

# If the regular expressions did not hit on any location information, pass it UNKN for now
for i in range(len(remark_uas_loc_df)):
    if len(uas_loc[i]) == 0:
        uas_loc[i] = 'UNKN'
```

```python
df1 = remark_uas_loc_df[remark_uas_loc_df['UAS Location'] != 'UNKN'].reset_index()
df1
```
