# Major Venues Exploratory Analysis

### Import packages

```python
import pandas as pd
import numpy as np
import math 
```

### Read in the csv file

```python
venues = pd.read_csv('C:/Users/Olga/Desktop/GMU/DAEN690/LEGO/Major_Sport_Venues.csv')
venues
```

### Check for NULL values

```python
venues.isnull().sum()
```

### Drop all venues that are not in the US

```python
venues = venues.loc[venues["COUNTRY"] == 'USA']
venues
```

### Keep needed columns

```python
venues = venues[['NAME', 'LATITUDE', 'LONGITUDE']]
venues
```

### Add two new columns for the converted Latitude and Longitude Values

```python
venues['LAT'] = ''
venues['LONG'] = ''
venues
```

### Convert the decimal degrees (DD) to degrees, minutes, seconds (DMS)

```python
for index, row in venues.iterrows():
    if row['LATITUDE'] > 0:
        hemi = 'N'
    elif row['LATITUDE'] < 0:
        hemi = 'S'
        
    D_lat = int(abs(math.modf(row['LATITUDE'])[1]))
    M_lat = int(abs(math.modf(math.modf(row['LATITUDE'])[0]*60)[1]))
    S_lat = abs(math.modf(math.modf(row['LATITUDE'])[0]*60)[0]*60)
    S_lat = round(S_lat, 3)    
    
    venues.loc[index, 'LAT'] = str(D_lat) + u"\u00b0 " + str(M_lat) + "' " + str(S_lat) + "\" " + hemi
    
for index, row in venues.iterrows():
    if row['LONGITUDE'] > 0:
        hemi = 'E'
    elif row['LONGITUDE'] < 0:
        hemi = 'W'
        
    D_long = int(abs(math.modf(row['LONGITUDE'])[1]))
    M_long = int(abs(math.modf(math.modf(row['LONGITUDE'])[0]*60)[1]))
    S_long = abs(math.modf(math.modf(row['LONGITUDE'])[0]*60)[0]*60)
    S_long = round(S_long, 3)    
    
    venues.loc[index, 'LONG'] = str(D_long) + u"\u00b0 " + str(M_long) + "' " + str(S_long) + "\" " + hemi
    

venues
```
